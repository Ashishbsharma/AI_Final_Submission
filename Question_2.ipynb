```python
import json

def get_valid_moves(board, player):
    valid_moves = []
    for i in range(3):
        for j in range(3):
            if board[i][j] == '-':
                new_board = [row.copy() for row in board]
                new_board[i][j] = player
                valid_moves.append(new_board)
    return valid_moves

def generate_states():
    initial_state = [['-' for _ in range(3)] for _ in range(3)]
    states = set()
    stack = [(initial_state, 'X')]

    while stack:
        current_state, current_player = stack.pop()

        if current_player == 'X':
            next_player = 'O'
        else:
            next_player = 'X'

        valid_moves = get_valid_moves(current_state, current_player)
        for move in valid_moves:
            state_tuple = tuple(tuple(row) for row in move)
            if state_tuple not in states:
                states.add(state_tuple)
                stack.append((move, next_player))

    return list(states)

def print_states(states):
    for state in states:
        state_str = ''
        for row in state:
            state_str += '|' + '|'.join(row) + '|\n'
        print(state_str)

if __name__ == "__main__":
    all_states = generate_states()
    print_states(all_states)

    with open('tic_tac_toe_states.json', 'w') as json_file:
        json.dump(all_states, json_file, indent=2)
```
